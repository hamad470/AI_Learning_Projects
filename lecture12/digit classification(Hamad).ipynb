```python
import pandas as pd 
```

```python
df = pd.read_csv("datasets/train.csv")
df
```

```python
y= df["label"]
X = df.drop(columns=["label"])
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier(criterion = "entropy")
```

```python
clf = clf.fit(X,y)
```

```python
import matplotlib.pyplot as plt

# Example of limiting the depth to 3
fig, ax = plt.subplots(figsize=(38, 28))  # Adjust the size as needed
tree.plot_tree(clf, max_depth=1, feature_names=X.columns, filled=True, ax=ax)
plt.show()
```

```python
df_t = pd.read_csv("datasets/test.csv")
df_t
```

```python
preds = clf.predict(df_t)
preds
```

```python
import numpy as np
index = 2  
df_t = np.array(df_t)
image = df_t[index].reshape(28, 28)


plt.imshow(image, cmap='gray')
plt.title(f"Digit at index {index}")
plt.axis('off') 
plt.show()
```

```python
preds = pd.DataFrame(preds)
preds.to_csv("datasets/predictions.csv")
```
