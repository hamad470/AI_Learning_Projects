```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df_X,df['species'],random_state =42,test_size = 0.2)
```

```python

zeros=[]
ones=[]
twos=[]

for i in y_train.index:
    if df["species"].loc[i] == 0:
        zeros.append(i)
    elif df["species"].loc[i] == 1:
        ones.append(i)
    else:
        twos.append(i)
print(zeros)
print(ones)
print(twos)
```
