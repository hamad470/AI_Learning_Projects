```python
df3["Last_Name"]=df3["Last_Name"].str.strip("./-")
df3
```

```python
df3["Phone_Number"]=df3["Phone_Number"].str.replace("N/a","")
df3
```

```python
#removing the rows in phone number having NaN and empty string
df3.dropna(subset="Phone_Number",inplace=True)
df3
```

```python
df3
```

```python
for i in df3.index:
    if df3.loc[i,"Phone_Number"] == "":
        df3.drop(i,inplace=True)
```

```python
df3
```

```python

df3["Phone_Number"]=df3["Phone_Number"].str.replace(r"[^a-zA-Z0-9]","",regex=True)
df3["Phone_Number"]=df3["Phone_Number"].apply(lambda x:str(x))
df3["Phone_Number"]=df3["Phone_Number"].apply(lambda x:x[0:3]+"-"+x[3:6]+"-"+x[6:10])
df3
```

```python

df3.drop(columns="Not_Useful_Column",inplace=True)
df3
```

```python
df3["Do_Not_Contact"]=df3["Do_Not_Contact"].apply(lambda x:str(x))
df3["Do_Not_Contact"]=df3["Do_Not_Contact"].apply(lambda x:x[0:1])
df3
```

```python
df3["Paying Customer"]=df3["Paying Customer"].apply(lambda x:x[0:1])
df3
```

```python
for i in df3.index:
    if df3.loc[i,"Do_Not_Contact"] == "Y":
        df3.drop(i,inplace = True)
df3
```

```python
#if you want to split Adress into two parts
#we needed to convert addresses to string types, because they were not
df3[["StreetAddress","City"]] = df3["Address"].astype(str).str.split(",", n=1,expand=True)
df3
```

```python
df3.drop(columns="Address",inplace = True)
df3
```
