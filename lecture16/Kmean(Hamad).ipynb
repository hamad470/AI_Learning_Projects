```python
import pandas as pd
df = pd.read_csv("student_clustering.csv")
```

```python
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    km = KMeans(n_clusters = i)
    km.fit_predict(df)
    wcss.append(km.inertia_)
plt.plot(range(1,11),wcss)
    
```

```python
km = KMeans(n_clusters = 4)
y_means = km.fit_predict(df)
```

```python
X=df.iloc[:,:].values
# converting  to numoy arrays
```

```python
plt.scatter(X[y_means==0,0],X[y_means==0,1],color = "blue")
plt.scatter(X[y_means==1,0],X[y_means==1,1],color = "green")
plt.scatter(X[y_means==2,0],X[y_means==2,1],color = "red")
plt.scatter(X[y_means==3,0],X[y_means==3,1],color = "black")
```

```python
# on 3d data
from sklearn.datasets import make_blobs
centroids = [(-5,-5,5),(5,5,-5),(3.5,-2.5,4),(-2.5,2.5,-4)]
cluster_std = [1,1,1,1]
X,y = make_blobs(n_samples  = 200,centers =centroids,cluster_std =cluster_std,random_state =1)
```

```python
X
```

```python
import plotly.express as px
```

```python
fig = px.scatter_3d(x=X[:,0],y=X[:,1],z=X[:,2])
fig.show()
```

```python
wcss =[]
for i in range(1,21):
    km = KMeans(n_clusters=i)
    km.fit_predict(X)
    wcss.append(km.inertia_)
plt.plot(range(1,21),wcss)
```

```python
km = KMeans(n_clusters=4)
y_preds = km.fit_predict(X)
```

```python
df["col1"] = X[:,0]
df["col2"] = X[:,1]
df["col3"] = X[:,2]
df["label"] = y_preds
fig = px.scatter_3d(df,x="col1",y="col2",z="col3",color="label")
fig.show()
```
