```python
def calculate_mean_value(value):
    if "-" in value:
        parts=value.split("-")
        if len(parts) ==2 and parts[0] != "":
        
            # min_val,max_val = map(float,value.split("-"))
            min_val,max_val = [float(i) for i in parts]
            # In Python, float is a built-in function that converts a string or a number to a floating-point number. It is used in the map function without parentheses because float is being passed as an argument to map, not being called directly.
            return (min_val+max_val)/2
        else:
            return float(value)
        
    else:
        return float(value)
```
