```python
df["Caliber"]=df["Caliber"].astype(str).str.replace(r"[^0-9]","",regex=True)
df["Caliber"]
```

```python
# df["Grip Length"] = df["Grip Length"].fillna((df["Grip Length"].astype(float).mean()))
# df["Grip Length"].isna().sum()
df["Grip Length"]=pd.to_numeric(df["Grip Length"],errors="coerce")
df["Grip Length"] =df["Grip Length"].fillna(df["Grip Length"].mean())
df["Grip Length"]
```
