```python
from bs4 import BeautifulSoup
import requests
```

```python
url  = "https://en.wikipedia.org/wiki/List_of_universities_in_Pakistan"
response = requests.get(url=url)

print(response.content)
```

```python
from collections import defaultdict
import pandas as pd
soup = BeautifulSoup(response.content , "html.parser")
tables = soup.find_all('table')
data = []
data2= []
theads = []
dataFrame = defaultdict(list)

# print("TOTAL TABLES :",len(tables))
# print(tables[0])
t_body = tables[0].find('tbody')
# print(t_body)
trows = t_body.find_all('tr')
cnt = 0
for tr in trows:
    cnt +=1
    ths = tr.find_all('th')
    for th in ths:

        theads.append(th.text)
    if cnt ==1:
        print("ENTER")
        continue

    tds = tr.find_all('td')


    count = 0     
    for td in tds:
        data.append(td.text)

    # print(data)
    data2.append(data.copy())
    data.clear()   
    count +=1 
        

print(len(data2))

# print(data2)

for data in data2:
    for  index , item in enumerate(data):
        dataFrame[theads[index]].append(item)


# Ensure all lists are of the same length by appending None or empty strings
max_length = max(len(lst) for lst in dataFrame.values())
for key in dataFrame:
    while len(dataFrame[key]) < max_length:
        dataFrame[key].append(None)  # Or use '' if you prefer empty strings

# Convert defaultdict to regular dict
dataFrame = dict(dataFrame)

# Convert dictionary to pandas DataFrame
df = pd.DataFrame(dataFrame)

# Save DataFrame to CSV
df.to_csv('output.csv', index=False)

print(df)


```
